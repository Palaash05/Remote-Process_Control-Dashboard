```python
import mysql.connector
from time import sleep
```

```python
# CONNECTING TO THE AWS RDS DATABASE

connection = mysql.connector.connect(
    host = 'HOST-NAME',
    port = PORT-NUMBER,
    user = 'USER-NAME',
    password = 'PASSWORD',
    database= 'DATABASE-NAME'
    )
```

```python
mycursor = connection.cursor()
```

### Creating Data

```python
import numpy as np
from scipy.stats import truncnorm
```

```python
# USING THIS FUNCTION TO CREATE NORMALLY DISTRIBUTED DATA
def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
```

### Banbury

```python
mycursor.execute("SHOW COLUMNS FROM scout.Banbury")
for x in mycursor:
    print(x)
```

```python
# DUMMY DATA 
banbury_width = get_truncated_normal(mean=20, sd=4, low=16, upp=26).rvs(200).astype(int)
```

```python
banbury_width
```

```python
# UPDATING DATA OF BANBURY IN DATABASE WITH  GAP OF 10 SECS IN BETWEEN EVERY UPDATE
for i in banbury_width:
    sleep(10)
    sql = """INSERT INTO Banbury (`sheet_width`) VALUES (%s)"""
    val = (int(i),)
    mycursor.execute(sql,val)
    connection.commit()
    print('executed {}'.format(i))
```

```python
mycursor.execute("SELECT * FROM Banbury")
```

```python
myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
